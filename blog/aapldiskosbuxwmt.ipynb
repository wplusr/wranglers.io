# brands i use everyday ðŸ¤ 
i decided to analyze the stocks of companies i interact with almost daily, becuase their brands are basically everywhere and have such a huge reach. my top five are apple, coca-cola, disney, starbucks and walmart.

```python
import yfinance as yf

# Define the tickers for the companies
tickers = ["AAPL", "KO", "DIS", "SBUX", "WMT"]

# Download the data
data = yf.download(tickers, period="5y")

# Display the first few rows of the data
display(data.head())
```

```python
import matplotlib.pyplot as plt

# Select the 'Close' price data
closing_prices = data['Close']

# Plot the closing prices
plt.figure(figsize=(12, 6))
for ticker in closing_prices.columns:
    plt.plot(closing_prices.index, closing_prices[ticker], label=ticker)

plt.title('Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Illustrative data for Starbucks' total revenue (for a recent fiscal year, e.g., FY2024).
# We sum up the segments to get a total for a simpler comparison.
# *** IMPORTANT: You should update these numbers with actual, recent figures from Starbucks' official financial reports (e.g., their latest 10-K filing) ***
starbucks_total_revenue = 29.76 + 4.51 + 1.9 # Sum of Company-Owned, Licensed, and Other
starbucks_data = {
    'Company': ['Starbucks'],
    'Revenue (Billions USD)': [starbucks_total_revenue]
}
df_starbucks_total = pd.DataFrame(starbucks_data)

# Illustrative data for Coca-Cola's total revenue (for a recent fiscal year, e.g., FY2024).
# *** IMPORTANT: You should update this number with actual, recent total revenue for Coca-Cola ***
coca_cola_data = {
    'Company': ['Coca-Cola'],
    'Revenue (Billions USD)': [47.0] # This is an illustrative total revenue figure
}
df_coca_cola_total = pd.DataFrame(coca_cola_data)

# Combine the DataFrames for both companies into a single DataFrame for direct comparison
df_combined_simple = pd.concat([df_starbucks_total, df_coca_cola_total], ignore_index=True)

# Set a nice style for the plot to make it visually appealing
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6)) # Adjust size for a simpler comparison

# Create a simple bar chart comparing total revenue side-by-side
# 'x' will be the Company, 'y' will be the total Revenue
sns.barplot(x='Company', y='Revenue (Billions USD)', data=df_combined_simple, palette=['#00704A', '#E4002B']) # Starbucks green, Coca-Cola red

# Add a clear title to the chart
plt.title('Starbucks vs. Coca-Cola: Illustrative Total Revenue Comparison', fontsize=16)

# Label the axes for clarity
plt.xlabel('Company', fontsize=12)
plt.ylabel('Total Revenue (Billions USD)', fontsize=12)

# Add the exact revenue values on top of each bar for precise reading
for p in plt.gca().patches:
    height = p.get_height()
    if height > 0:
        plt.text(p.get_x() + p.get_width()/2., height,
                 f'${height:.2f}B',
                 ha='center', va='bottom', fontsize=10)

plt.show()
```
