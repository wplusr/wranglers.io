## Fetch stock data

### Subtask:
Retrieve historical stock data for Uber and Lyft using a financial data library.


**Reasoning**:
Import the `yfinance` library and download the historical stock data for Uber and Lyft using their ticker symbols and a relevant date range, storing the data in separate dataframes.



```python
import yfinance as yf

# Define the ticker symbols for Uber and Lyft
uber_ticker = 'UBER'
lyft_ticker = 'LYFT'

# Define the date range for the analysis (from their IPO dates to present)
start_date = '2019-05-10' # Uber IPO date
end_date = '2025-07-05'

# Fetch historical stock data for Uber and Lyft
uber_data = yf.download(uber_ticker, start=start_date, end=end_date)
lyft_data = yf.download(lyft_ticker, start=start_date, end=end_date)

# Display the first few rows of the dataframes
display(uber_data.head())
display(lyft_data.head())
```

## Analyze data and identify key periods

### Subtask:
Analyze the stock data to identify periods of significant highs and lows for both companies.


**Reasoning**:
Find the all-time highest and lowest 'Close' prices and their dates for both Uber and Lyft and store them.



```python
uber_highest_close = uber_data['Close'].max()
uber_highest_close_date = uber_data['Close'].idxmax()
uber_lowest_close = uber_data['Close'].min()
uber_lowest_close_date = uber_data['Close'].idxmin()

lyft_highest_close = lyft_data['Close'].max()
lyft_highest_close_date = lyft_data['Close'].idxmax()
lyft_lowest_close = lyft_data['Close'].min()
lyft_lowest_close_date = lyft_data['Close'].idxmin()

print("Uber All-Time Highest Close:")
print(f"Date: {uber_highest_close_date}, Price: {uber_highest_close}")
print("\nUber All-Time Lowest Close:")
print(f"Date: {uber_lowest_close_date}, Price: {uber_lowest_close}")

print("\nLyft All-Time Highest Close:")
print(f"Date: {lyft_highest_close_date}, Price: {lyft_highest_close}")
print("\nLyft All-Time Lowest Close:")
print(f"Date: {lyft_lowest_close_date}, Price: {lyft_lowest_close}")
```

## Generate visualizations

### Subtask:
Create visualizations (e.g., line plots) of the stock prices for Uber and Lyft to illustrate their performance over time.


**Reasoning**:
Create line plots for Uber and Lyft stock prices over time as instructed.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(uber_data.index, uber_data['Close'], label='Uber')
plt.plot(lyft_data.index, lyft_data['Close'], label='Lyft')
plt.title('Uber and Lyft Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
